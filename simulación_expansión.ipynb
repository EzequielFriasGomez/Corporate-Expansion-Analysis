# Análisis de Rentabilidad y Expansión Corporativa
**Objetivo:** Transformación, localización e ingeniería de características de un dataset comercial estadounidense. El proceso incluye la traducción de variables al español, limpieza de columnas irrelevantes y el cálculo de costos operativos para permitir un análisis de márgenes en Power BI.

```python
import pandas as pd

# Cargamos el dataset crudo
df = pd.read_csv("DS Original.csv")
```

### 1. Limpieza Estructural y Traducción de Columnas
Se eliminan variables que no aportan al análisis regional y categórico (como Row ID o Nombres de clientes). Posteriormente, se estandariza el nombre de las columnas al español.

```python
# Eliminamos las columnas que no se exportaron en tu versión final
columnas_basura = ["Row ID", "Ship Date", "Ship Mode", "Customer Name", "Country"]
df = df.drop(columns=columnas_basura)

# Diccionario de traducción de columnas
renombres = {
    "Order ID": "ID_Pedido",
    "Order Date": "Fecha_Pedido",
    "Segment": "Segmento",
    "City": "Ciudad",
    "State": "Estado",
    "Postal Code": "Codigo_Postal",
    "Category": "Categoria",
    "Sub-Category": "Sub_Categoria",
    "Product Name": "Nombre_Producto",
    "Sales": "Ventas",
    "Quantity": "Cantidad",
    "Discount": "Descuento",
    "Profit": "Ganancia"
}
df = df.rename(columns=renombres)

# Convertimos la fecha al formato estándar (YYYY-MM-DD)
df["Fecha_Pedido"] = pd.to_datetime(df["Fecha_Pedido"])
```

### 2. Localización de Datos (Mapeo)
Traducción de los valores internos de las columnas clave para que el tablero final sea comprensible para la gerencia hispanohablante.

```python
# Diccionarios de traducción de valores
dic_segmento = {"Consumer": "Consumidor Final", "Corporate": "Corporativo", "Home Office": "Oficina en Casa"}
dic_region = {"South": "Sur", "West": "Oeste", "Central": "Centro", "East": "Este"}
dic_categoria = {"Furniture": "Muebles", "Office Supplies": "Insumos de Oficina", "Technology": "Tecnología"}

# Aplicamos el mapeo (fillna mantiene el valor original si no encuentra traducción, ideal para subcategorías muy largas)
df["Segmento"] = df["Segmento"].map(dic_segmento).fillna(df["Segmento"])
df["Region"] = df["Region"].map(dic_region).fillna(df["Region"])
df["Categoria"] = df["Categoria"].map(dic_categoria).fillna(df["Categoria"])

# Nota: Podés sumar al código un diccionario para "Sub_Categoria" si querés traducir "Bookcases" a "Bibliotecas", etc.
```

### 3. Ingeniería de Características: Costo Operativo
El dataset original no provee el costo de los productos. Para poder auditar la rentabilidad, calculamos esta métrica deduciendo la ganancia del volumen de ventas total.

```python
# Creación de la columna Costo
df["Costo"] = df["Ventas"] - df["Ganancia"]
```

### 4. Exportación del Modelo

```python
df.to_csv("Simulación_Expansión.csv", index=False)
```
